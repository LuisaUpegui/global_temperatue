```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Colab Notebooks/Coding dojo/global temperature/Global Temperature Anomalies.csv"
df = pd.read_csv(filename)
df
```

```python
year = df['Hemisphere'] == "Global"
fil = df.loc[year, :].groupby(['Year'])['J-D'].mean().sort_values(ascending = False)
fil
```

```python
fil
```

```python
# This shows how the the temperature anomalies has increasing considerably since 1977. 
# temperature since 1977 to 2020  has increased almost three times.


sns.barplot(x = fil.index, y = fil.values, palette="OrRd").set(title= "Global average surface temperature");
plt.xticks(rotation = 90,fontsize = 7  );
sns.set(rc= {'figure.figsize':(15,8)})
```
